```python
data['quality_label'] = data['quality'].apply(lambda x: 'Good Quality' if x > 6 else 'Bad Quality')

# Show some examples of good quality and bad quality wines
good_quality_wines = data[data['quality_label'] == 'Good Quality'].head(5)
bad_quality_wines = data[data['quality_label'] == 'Bad Quality'].head(5)

print("Good Quality Wines:")
print(good_quality_wines)
print("\nBad Quality Wines:")
print(bad_quality_wines)
```

```python
x = data.drop(['quality','quality_label'], axis=1)
y= data['quality_label']
```

```python

### now we have to split our dataset
## we have to first train our machin then test from dataset

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=1234)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

```

# Train Model : Logistic Rregression

```python
### Now train our model 
from sklearn.linear_model import LogisticRegression
```

```python
##Fit a model to the training data.Using regression model as my task suggested
log = LogisticRegression(solver='lbfgs', max_iter=1000)
log.fit(x_train,y_train)

```

```python
## Predict and Evaluate:
##Use the trained model to make predictions on the test data and evaluate performance
from sklearn.metrics import  accuracy_score , classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')


log_pred = log.predict(x_test)

y_test_str = y_test.astype(str)
log_pred_str = log_pred.astype(str)
## predict accuracy

accuracy = accuracy_score(y_test_str, log_pred_str)
print('Accuracy:', accuracy)
print('Classification Report:')
print(classification_report(y_test_str, log_pred_str))
print('Confusion Matrix:')
print(confusion_matrix(y_test_str, log_pred_str))


```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier


# Train the model
DTC= DecisionTreeClassifier()
DTC.fit(x_train, y_train)
```

```python
from sklearn.metrics import  accuracy_score , classification_report, confusion_matrix

DTC_pred = DTC.predict(x_test)

## predict accuracy
accuracy = accuracy_score(y_test, DTC_pred)
print('Accuracy:', accuracy)
print('Classification Report:')
print(classification_report(y_test, DTC_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, DTC_pred))
```

# PLOT Confusion matrix

```python
# Plotting the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, DTC_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier


# Train the model
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
```

```python
from sklearn.metrics import  accuracy_score , classification_report, confusion_matrix

RFC_pred = RFC.predict(x_test)
## predict accuracy

accuracy = accuracy_score(y_test, RFC_pred)
print('Accuracy:', accuracy)
print('Classification Report:')
print(classification_report(y_test, RFC_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, RFC_pred))
```

# PLOT Confusion matrix

```python
# Plotting the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, RFC_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
